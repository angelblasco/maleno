```python
# Aproximacion de distribucion de Poisson y Normal a la distribucion Binomial.
```

```python
# Consideraremos los parametros n=200 y p=0.5
```

```python
n = 200
p = 0.5
```

```python
import numpy as np
import scipy.stats as ss
import matplotlib.pyplot as plt
```

```python
# Array con el numero de exitos 
```

```python
k = np.arange(100)
```

```python
k
```

```python
ybinomial = ss.binom(n,p)
```

```python
binomial = ybinomial.cdf(k) # cdf carga la funcion de distribucion. Viene en scipy.stats
```

```python
plt.plot(k,binomial)
plt.show()
    
```

```python
ypoisson = ss.poisson(n*p)
```

```python
poison = ypoisson.cdf(k)
```

```python
plt.plot(k,poison)
plt.show()
```

```python
# Ahora comparamos ambas graficas
```

```python
plt.plot(k,binomial,k,poison)
plt.show()
```
