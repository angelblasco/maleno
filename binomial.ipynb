```python
# Aproximacion de distribucion de Poisson y Normal a la distribucion Binomial.
```

```python
# Consideraremos los parametros n="numero de repeticiones" y p="probabilidad de exito".
```

```python
n = 30
p = 0.033
```

```python
import numpy as np
import scipy.stats as ss
import matplotlib.pyplot as plt
```

```python
# Deshabilitamos la notacion cientifica.
```

```python
np.set_printoptions(precision = 5, suppress = True)
```

```python
# Array con el numero de exitos. 
```

```python
k = np.arange(11)
```

```python
k
```

```python
ybinomial = ss.binom(n,p)
```

```python
binomial = ybinomial.pmf(k) # pmf carga la funcion de masa de la distribucion. Viene en scipy.stats
```

```python
plt.plot(k,binomial)
plt.xlabel("Numero de exitos")
plt.ylabel("Probabilidad")
plt.grid(True)
plt.show()
```

```python
ypoisson = ss.poisson(n*p)
```

```python
poison = ypoisson.pmf(k)
```

```python
plt.plot(k,poison)
plt.xlabel("Numero de exitos")
plt.ylabel("Probabilidad")
plt.grid(True)
plt.show()
```

```python
# Ahora comparamos ambas graficas
```

```python
plt.plot(k,binomial,label="Binomial")
plt.plot(k,poison,label="Poisson")
plt.xlabel("Numero de exitos")
plt.ylabel("Probabilidad")
plt.grid(True)
plt.legend()
plt.show()
```
