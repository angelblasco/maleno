```python
%matplotlib inline
```

```python
n = 400
p = 0.0333
```

```python
import numpy as np
import scipy.stats as ss
import matplotlib.pyplot as plt
```

```python
np.set_printoptions(precision = 5, suppress = True)
```

```python
k = np.arange(31)
```

```python
ybinomial = ss.binom(n,p)
```

```python
binomial = ybinomial.pmf(k) # pmf carga la funcion de masa de la distribucion. Viene en scipy.stats
```

```python
plt.figure(figsize=(10,7))
plt.plot(k,binomial,label="Binomial")
plt.xlabel("Numero de exitos")
plt.ylabel("Probabilidad")
plt.grid(True)
plt.legend()
plt.show()
```

```python
ypoisson = ss.poisson(n*p)
```

```python
poison = ypoisson.pmf(k)
```

```python
plt.figure(figsize=(10,7))
plt.plot(k,poison,label="Poisson")
plt.xlabel("Numero de exitos")
plt.ylabel("Probabilidad")
plt.grid(True)
plt.legend()
plt.show()
```

```python
plt.figure(figsize=(10,7))
plt.title("Comparacion entre Binomial y Poisson")
plt.plot(k,binomial,label="Binomial")
plt.plot(k,poison,label="Poisson")
plt.xlabel("Numero de exitos")
plt.ylabel("Probabilidad")
plt.grid(True)
plt.legend()
plt.show()
```
